```python
data = [
    "Film menggugah pikiran, sangat nasionalis #OneForAll",
    "Sangat dramatis, emosinya dapat, terutama emosi penonton",
    "Menyesal beli tiket",
    "Saya suka film ini. Saya ajak satu keluarga sampe nangis semua"
]
```

```python
for row in data:
  print("row:", row)
  prompt = f"""
  Please classify this review to either positive or negative.
  Respond with label only

  {row}
  """
  print("prompt:", prompt)

  output = llm.invoke(prompt)
  print("output:", output)

  print("=========================")
```

```python
import pandas as pd

df = pd.read_csv("powerconsumption.csv")
df
```

```python
from langchain_experimental.agents.agent_toolkits.pandas.base import (
    create_pandas_dataframe_agent
)

agent = create_pandas_dataframe_agent(
    llm=Replicate(model="ibm-granite/granite-3.3-8b-instruct"),
    df=df,
    verbose=True,
    allow_dangerous_code=True,  # opt-in for Python REPL execution
)
```

```python
agent.invoke({"input": "under what conditions to get the ideal Power Consumption is in the dataset and give the time and value"})
```

```python
agent.invoke({"input": "make me a comparison chart between the best values ​​by determining the temperature sensor and other sensors to produce the ideal power consumption from the data sheet"})
```
